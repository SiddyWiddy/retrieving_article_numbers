```python
import numpy as np
from datetime import date #importing the date module
from bs4 import BeautifulSoup #importing web scraping module
import pandas as pd # importing the database module
import requests #importing the GET requests for web scraping module
import matplotlib.pyplot as plt #importing graph module
years_Articles = {} #creating an empty dictionary 
current_date = date.today() #getting the local date today
n = int(current_date.year) #getting the year from the local date
def getYearResults(ur): #creating a function called getYearResults -> This is the web scraping for articles function
    webpage = requests.get(ur)
    sou = BeautifulSoup(webpage.content)
    links=sou.find_all('div',attrs={'class':'gs_ab_mdw'})
    results = links[1]
    temp = results.text.split("(")[0].split(" ")[1]
    num = temp.split(",")
    num=int(''.join(num))
    return (num)
element = input("""Here are choices for battery element studies:
Lithium, Sodium, Calcium, Magnesium, lithium-Sulphur, Sodium-Air, Lithium Oxide, Sodium Oxide, Sodium-Sulphur
Aluminium, or Potassium: """) #asks the user to input in the study they want
desired_year = int(input("Enter which year do you want the number of articles from: ")) #asks the user to enter the year they want it from
for i in range(desired_year,n): #loops through all the years, starting from the input to the present date.
    if element == "Lithium" or element == "Li" or element == "li": #manages Lithium output
        lin = getYearResults(("https://scholar.google.com/scholar?q=lithium+ion+battery&hl=en&as_sdt=0%2C5&as_ylo={}&as_yhi={}".format(i,i)))
        years_Articles[i] = lin #stores the years and article number in dictionary
    elif element == "Sodium" or element == "Na" or element == "na":#manages Sodium output
        lin2 = getYearResults(("https://scholar.google.com/scholar?q=sodium+ion+battery&hl=en&as_sdt=0%2C5&as_ylo={}&as_yhi={}".format(i,i)))
        years_Articles[i] = lin2 #stores the years and article number in dictionary
    elif element == "Calcium" or element == "Ca" or element == "ca":#manages Calcium output
        lin3 = getYearResults(("https://scholar.google.com/scholar?q=calcium+ion+battery&hl=en&as_sdt=0%2C5&as_ylo={}&as_yhi={}".format(i,i)))
        years_Articles[i] = lin3 #stores the years and article number in dictionary
    elif element == "Magensium" or element == "Mg" or element == "mg":#manages Magnesium output
        lin4 = getYearResults(("https://scholar.google.com/scholar?q=Magnesium+ion+battery&hl=en&as_sdt=0%2C5&as_ylo={}&as_yhi={}".format(i,i)))
        years_Articles[i] = lin4 #stores the years and article number in dictionary
    elif element == "Lithium-Sulphur" or element == "Li-S":#manages Lithium-Sulphur output
        lin5 = getYearResults(("https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&as_ylo={}&as_yhi={}&q=lithium+sulphur+battery+&btnG=".format(i,i)))
        years_Articles[i] = lin5 #stores the years and article number in dictionary
    elif element == "Sodium-Air" or element == "Na-Air":#manages Sodium-Air output
        lin6 = getYearResults(("https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&as_ylo={}&as_yhi={}&q=Sodium+air+battery&btnG=".format(i,i)))
        years_Articles[i] = lin6 #stores the years and article number in dictionary
    elif element == "Lithium Oxide" or element == "LI-O":
        lin7 =getYearResults(("https://scholar.google.com/scholar?q=Lithium+oxide+battery&hl=en&as_sdt=0%2C5&as_ylo={}&as_yhi={}".format(i,i)))
        years_Articles[i] = lin7
    elif element == "Sodium Oxide" or element == "Na-O":
        lin8 = getYearResults(("https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&as_ylo={}&as_yhi={}&q=Sodium+oxide+battery&btnG=".format(i,i)))
        years_Articles[i] = lin8
    elif element == "Sodium-Sulphur" or element == "Na-S" or element == "sodium-sulphur" or element == "na-s":
        lin9 = getYearResults(("https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&as_ylo={}&as_yhi={}&q=Sodium+Sulphur+battery&btnG=".format(i,i)))
        years_Articles[i] = lin9
    elif element == "Aluminium" or element == "Al" or element == "aluminium" or element == "al":
        lin10 = getYearResults(("https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&as_ylo={}&as_yhi=&q=Aluminium+ion+battery&btnG=".format(i,i)))
        years_Articles[i] = lin10
    elif element == "Potassium" or element == "K" or element == 'potassium' or element == "k":
        lin11 = getYearResults(("https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&as_ylo={}&as_yhi={}&q=Potassium+ion+battery&btnG=".format(i,i)))
        years_Articles[i] = lin11
data_items = years_Articles.items() 
data_lst = list(data_items)
df = pd.DataFrame(data_lst)
df.columns=['year','Articles']
#plt.plot((*zip(*sorted(years_Articles.items()))),color="red", label="Graph Results")
key = np.fromiter(years_Articles.keys(), dtype=float)
val = np.fromiter(years_Articles.values(), dtype=float)
plt.scatter(key,val,color="b", label="Graph Results",s=20*4**0.5)#,s = 20*4**3)
m, c = np.polyfit(key, val, 1)
plt.plot(key, m*key + c, label="Line of Best fit",lw = 3,color="red",alpha=0.85)
plt.xlabel("Year")
plt.ylabel("Number of articles")
plt.legend(loc='lower right')
plt.show() #plotted graph
print("The gradient for the trend - line of best fit is: " + str(m))
df #drawn database
```
